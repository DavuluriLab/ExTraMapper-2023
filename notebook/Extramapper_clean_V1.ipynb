```python
#create a list of orthologous genes for input into summary code
import os

# Specify the directory you want to search
directory = "/content/ExTraMapper/Human-Mouse-Preprocess-Data/output/human-mouse"

# Specify the output file name
output_file = "gene-pair.list"

# Get all subdirectories within the specified directory
subdirectories = [subdir for subdir in os.listdir(directory) if os.path.isdir(os.path.join(directory, subdir))]

```

```python
# Write the subdirectories to the output file
with open(output_file, 'w') as file:
    file.write('\n'.join(subdirectories))

print(f"Subdirectories list has been saved to {output_file}.")
```
